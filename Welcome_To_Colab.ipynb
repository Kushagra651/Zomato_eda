```
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```
ratings
```

whenever the rating is between 4.5 --->4.9 the food is excellent<br>
whenever the rating is between 4.0 --->4.4 the food is very good

```
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',palette=['red','orange','gold','yellow','green'],data=ratings)
```

```
sns.countplot(x='Rating color',data =ratings,palette=['red','orange','gold','yellow','green'])
```
